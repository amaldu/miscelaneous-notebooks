# Import dependencies

```python
from kaggle.api.kaggle_api_extended import KaggleApi
import os
import pandas as pd
import glob

api = KaggleApi()
api.authenticate()
```

# View datasets

```python
datasets = api.dataset_list(search='uber', page=1)
print("Found datasets containing 'uber':")
for i, ds in enumerate(datasets):
    print(f"{i+1}. {ds.ref} - {ds.title}")
```

# Import Dataset

```python
dataset_ref = 'yashdevladdha/uber-ride-analytics-dashboard'
download_path = '../data/raw'

os.makedirs(download_path, exist_ok=True)
api.dataset_download_files(dataset_ref, path=download_path, unzip=True)
```

# Load Dataset

```python
csv_files = glob.glob(os.path.join(download_path, "*.csv"))
df = pd.read_csv(csv_files[0])
df
```
